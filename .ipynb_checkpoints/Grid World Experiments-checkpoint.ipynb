```python
# Imports

import numpy as np
import plotly.graph_objects as go
import random

import torch
import numpy as np
from collections import deque, namedtuple

import learning_agents
from cube_env import Cube,visualise
```

```python
class grid_nxn:
   
    def __init__(self,n):
        self.location = np.array([0,0])
        self.goal = np.array([n-1,n-1])
        self.n = n
        
    #Dynamics
    # 0:up 1:right 2:down 3:left
    def move(self,action):
        if action == 0:
            self.location[0] += 1
            if self.location[0] == self.n:
                self.location[0] = self.n-1
            return self.location
        if action == 1:
            self.location[1] += 1
            if self.location[1] == self.n:
                self.location[1] = self.n-1
            return self.location
        if action == 2:
            self.location[0] -= 1
            if self.location[0] == -1:
                self.location[0] = 0
            return self.location
        if action == 3:
            self.location[1] -= 1
            if self.location[1] == -1:
                self.location[1] = 0
            return self.location
        
    def checkReward(self):
        if self.checkDone():
            return +100
        return -1
    
    def checkDone(self):
        if (self.location == self.goal).all():
            return True
        return False
        
    def returnState(self):
        return self.location
    
    def reverseAction(self,action):
        if action == 0:
            return 2
        if action == 1:
            return 3
        if action == 2:
            return 0
        if action == 3:
            return 1
```

```python
# Defining DQN-EBU Algorithm

n = 5
grid = grid_nxn(n)

state_shape = grid.returnState().shape[0]
action_shape = 4



def dqn_ebu(n_episodes=100000, max_t=int(3*n), eps_start=1.0, eps_end=0.1, eps_decay=0.9995):
    print(pow(eps_decay,n_episodes))

    scores = []                 # list containing scores from each episode
    scores_window_printing = deque(maxlen=10) # For printing in the graph
    scores_window= deque(maxlen=100)  # last 100 scores for checking if the avg is more than 195
    eps = eps_start                    # initialize epsilon
    
    #Check if agent learns to solve a cube that is one move away from goal state
    
    
    for i_episode in range(1, n_episodes+1):
        
        grid = grid_nxn(n)
        state = grid.returnState()
        score = 0
        done = False
        for t in range(max_t):
            
            #Storing all steps in a list for the trajectory
            traj_val = []
            
            
            #Choosing an action
            action = agent.act(state, eps)

            #Executing that action
            grid.move(action)
            
            #Next state
            next_state = grid.returnState()
            reward = grid.checkReward()
            
            #Checking if the episode ended
            if grid.checkDone():
                done = True
                
            traj_val.append([state,action,reward,next_state,done])
                
            #agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            
            if done:
                break 
                
        for sublist in reversed(traj_val):
            agent.step(sublist[0], sublist[1], sublist[2], sublist[3], sublist[4])
            
        
        scores_window.append(score)                       # save most recent score
        scores_window_printing.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps)                 # decrease epsilon
        print('\rEpisode {}\tScore: {:.2f}'.format(i_episode, score), end="")        
        if i_episode % 100 == 0: 
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            scores.append(np.mean(scores_window))
        if np.mean(scores_window)>=100 - 2.5*n:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
            break
    return [np.array(scores),i_episode-100]

device = torch.device("mps" if torch.cuda.is_available() else "cpu")
agent = learning_agents.Agent_DQNEBU(state_size=state_shape,action_size = action_shape,seed = 0)
scores_ebu, terminal_ep_ebu = dqn_ebu()
```

```python
# Defining DQN-HER Algorithm

n = 10
grid = grid_nxn(n)

state_shape = grid.returnState().shape[0]
action_shape = 4



def dqn_her(n_episodes=100000, max_t=int(3*n), eps_start=1.0, eps_end=0.1, eps_decay=0.9995):
    print(pow(eps_decay,n_episodes))

    scores = []                 # list containing scores from each episode
    scores_window_printing = deque(maxlen=10) # For printing in the graph
    scores_window= deque(maxlen=100)  # last 100 scores for checking if the avg is more than 195
    eps = eps_start                    # initialize epsilon
    
    #Check if agent learns to solve a cube that is one move away from goal state
    
    
    for i_episode in range(1, n_episodes+1):
        
        grid = grid_nxn(n)
        fin_goal = grid.goal
        state = grid.returnState()
        score = 0
        done = False
         
        # Run for one trajectory
        for t in range(max_t):
            
            #We store all we need for each trajectory
            traj_val = []
            
            #Choosing an action
            action = agent.act(np.concatenate((grid.returnState(),fin_goal)), eps)

            #Executing that action
            grid.move(action)
            
            #Next state
            next_state = grid.returnState()
            reward = grid.checkReward()
            
            #Checking if the episode ended
            if grid.checkDone():
                done = True
                
            traj_val.append([state,action,reward,next_state,done])
            #agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            
            if done:
                break 
                
        # Once the trajectory is done, append the final state that the trajectory reached to the state and push it to experience replay
        psuedo_goal = next_state
        
        for sublist in traj_val:
            sublist[0] = np.concatenate((sublist[0],psuedo_goal))
            sublist[3] = np.concatenate((sublist[3],psuedo_goal))
            agent.step(sublist[0], sublist[1], sublist[2], sublist[3], sublist[4])
            
        
        
            
        scores_window.append(score)                       # save most recent score
        scores_window_printing.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps)                 # decrease epsilon
        print('\rEpisode {}\tScore: {:.2f}'.format(i_episode, score), end="")        
        if i_episode % 100 == 0: 
            
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            scores.append(np.mean(scores_window))
        if np.mean(scores_window)>=100 - 2.5*n:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
            break
    return [np.array(scores),i_episode-100]

device = torch.device("mps" if torch.cuda.is_available() else "cpu")
agent = learning_agents.Agent_DQNHER(state_size=state_shape,action_size = action_shape,seed = 0)
scores_her, terminal_ep_her = dqn_her()
```

```python
# Defining SRE-DQN with hindsight Algorithm

n = 10
grid = grid_nxn(n)

state_shape = grid.returnState().shape[0]
action_shape = 4



def dqn_sre(n_episodes=100000, max_t=int(3*n), eps_start=1.0, eps_end=0.1, eps_decay=0.9995):
    print(pow(eps_decay,n_episodes))

    scores = []                 # list containing scores from each episode
    scores_window_printing = deque(maxlen=10) # For printing in the graph
    scores_window= deque(maxlen=100)  # last 100 scores for checking if the avg is more than 195
    eps = eps_start                    # initialize epsilon
    
    #Check if agent learns to solve a cube that is one move away from goal state
    
    
    for i_episode in range(1, n_episodes+1):
        
        grid = grid_nxn(n)
        fin_goal = grid.goal
        state = grid.returnState()
        score = 0
        done = False
         
        # Run for one trajectory
        for t in range(max_t):
            
            #We store all we need for each trajectory
            traj_val = []
            
            #Choosing an action
            action = agent.act(np.concatenate((grid.returnState(),fin_goal)), eps)

            #Executing that action
            grid.move(action)
            
            #Next state
            next_state = grid.returnState()
            reward = grid.checkReward()
            
            #Checking if the episode ended
            if grid.checkDone():
                done = True
                
            traj_val.append([state,action,reward,next_state,done])
            #agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            
            if done:
                break 
                
        # Once the trajectory is done, append the final state that the trajectory reached to the state and push it to experience replay
        psuedo_goal = next_state
        
        for sublist in traj_val:
            sublist[0] = np.concatenate((sublist[0],psuedo_goal))
            sublist[3] = np.concatenate((sublist[3],psuedo_goal))
            agent.step(sublist[0], sublist[1], sublist[2], sublist[3], sublist[4])
        
        #For scramble trajectories
        
        state = grid.returnState()
        done = False
        
        scramble_length = int(2*n*random.random())
        grid.location = grid.goal
        for i in range(scramble_length):
            
            #Scrambler Part
            curr_state = grid.location
            reward = grid.checkReward()
            done = grid.checkDone()
            
            scramble_action = random.choice(np.arange(2,3))
            
            #if i == 0:
                #scramble_action = int(3+ 2*random.random())
            #else:
                #scramble_action = agent.scramble_action(curr_state,eps)
                
            rev_action = np.int32(grid.reverseAction(scramble_action))
                
            grid.move(scramble_action)
            
            next_state = grid.location
            
            agent.step(np.concatenate((next_state,fin_goal)), rev_action, reward, np.concatenate((curr_state,fin_goal)), done)
            
            if done:
                break
                
            #Repeater Part
            repeater_state = next_state
            
            for _ in range(i):
                
                #Setting the state for repeater to explore
                grid.location = repeater_state
                
                #Choosing an action
                rep_action = agent.agent_act(np.concatenate((grid.returnState(),fin_goal)), eps) 
                
                #Executing that action
                grid.move(rep_action)
                
                # Repeater Next state
                r_next_state = grid.returnState()
                rep_reward = grid.checkReward()
                
                if grid.checkDone():
                    done = True
                    
                agent.step(np.concatenate((repeater_state,fin_goal)), rep_action, rep_reward, np.concatenate((r_next_state,fin_goal)), done)
                
                
                

        

            
        
        scores_window.append(score)                       # save most recent score
        scores_window_printing.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps)                 # decrease epsilon
        print('\rEpisode {}\tScore: {:.2f}'.format(i_episode, score), end="")        
        if i_episode % 100 == 0: 
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            scores.append(np.mean(scores_window))
        if np.mean(scores_window)>=100 - 2.5*n:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
            break
    return [np.array(scores),i_episode-100]

device = torch.device("mps" if torch.cuda.is_available() else "cpu")
agent = learning_agents.Agent_DQNHER(state_size=state_shape,action_size = action_shape,seed = 0)
scores_sre, terminal_ep_sre = dqn_sre()
```
